```python
from requests_html import HTML,HTMLSession
import pandas as pd

session = HTMLSession()
```

```python
import openpyxl
```

```python
counter_index=0
```

```python
def get_Joes_company_details(joes_url):
    global counter_index
    dic1 = {}
    r = session.get(joes_url)
    company_name_address = r.html.find('.d3',first=True).text.split('\n')
    c_name = company_name_address[1]
    
    try:
        c_address = company_name_address[2]
    except:
        c_address = ''
        
    dic1['Company'] = c_name
    dic1['address'] = c_address

    company_details = r.html.find('.company_data',first=True).text.split('\n')
    counter_index = counter_index + 1
    
    for company in company_details:
        if company == '':
            continue
        company = company.split(':')
        dic1[company[0]] =company[1]
    company_frame = pd.DataFrame(dic1, index=[c_name])
    
    return company_frame
    
```

```python
#Read The Excel File From the Defined Path
path = "D:\\Python_Scraping_Scripts\\joesdata.xlsx"


wb_obj = openpyxl.load_workbook(path) 
sheet_obj = wb_obj.active 
m_row = sheet_obj.max_row 

cell_obj = sheet_obj.cell(row = 1, column = 2) 
company_dataFrame = get_Joes_company_details(cell_obj.value)

for i in range(2, m_row + 1): 
    cell_obj = sheet_obj.cell(row = i, column = 2) 
    print(cell_obj.value) 
    company_frame = get_Joes_company_details(cell_obj.value)
    company_dataFrame = pd.concat([company_dataFrame,company_frame])
    
```
